```python
import nltk
import matplotlib.pyplot as plt
%matplotlib inline
def word_analyze(bookStr, title):
    # simple lowercase tokenize
    tokens = nltk.word_tokenize(bookStr.lower())

    # filter out tokens that aren't words
    wordTokensLowerCase = [word for word in tokens if word[0].isalpha()]

    
    for n in range(2, 10):
        print("-----" + str(n) + "-grams in " + title + "-----")
        ngrams = list(nltk.ngrams(wordTokensLowerCase, n)) # create four-grams
        ngramsFreqs = nltk.FreqDist(ngrams) # determine frequency of four-grams
        for words, count in ngramsFreqs.most_common(10): # for the 15 most common four-grams
            print(" ".join(list(words)), "       ", count) # show the count and the create a string from the tuple
```

```python
import io

fullBookStr = ""
# read The books plain text into string
for i in range(1, 25):
    with io.open("data/"+ str(i), "r",encoding='utf8') as f:
        bookStr = f.read()
    fullBookStr += "\n" + bookStr
    #word_analyze(bookStr, "Book " + str(i))

word_analyze(fullBookStr, "The Iliad")
```
