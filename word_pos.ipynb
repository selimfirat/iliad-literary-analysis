```python
import nltk
import matplotlib.pyplot as plt
%matplotlib inline

def get_percentage_pos(string, pos):
    tokens = nltk.word_tokenize(string)
    tagged = nltk.pos_tag(tokens)
    matching_tagged = [word for word, p in tagged if pos == p]
    return len(matching_tagged) / len(tagged)

def pos_analyze(bookStr, title):
    # simple lowercase tokenize
    tokens = nltk.word_tokenize(bookStr.lower())
    tagged = nltk.pos_tag(tokens)
                          
    tagsWP = [pos for word, pos in tagged]
    tags = [pos if pos.isalpha() else "." for pos in tagsWP]
    tagsFreqs = nltk.FreqDist(tags)
    tagsFreqs.tabulate(25)
    tagsFreqs.plot(25, title="Tag Frequencies in " + title)
    
    
    print(title + " has", '{:.2%}'.format(get_percentage_pos(bookStr, "JJ")), "adjectives")
```

```python
import io

fullBookStr = ""
# read The books plain text into string
for i in range(1, 25):
    with io.open("data/"+ str(i), "r",encoding='utf8') as f:
        bookStr = f.read()
    fullBookStr += "\n" + bookStr
    pos_analyze(bookStr, "Book " + str(i))

pos_analyze(fullBookStr, "The Iliad")
```
